La ecuación de onda es:

$$
\frac{\partial^2 u}{\partial t^2} = c^2 \left(\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2}\right)
$$

con condiciones de frontera

$$
u=0
$$
y las condiciones iniciales
$$
u(x,y,0)=f(x,y)
$$
Velocidad inicial: $$\frac{\partial u}{\partial t}=g(x,y) $$
Este problema des de Sturm por lo que tenemos la solucion como: \\
Primer paso:
Tenemos que las dimensiones son a=4 y b=2
Resolvemos por el metodo de separacion de variables en el plano (x,y) deonde ahora tenemos las edos \\

Tenemos que nuestras condiciones iniciales son \\
H(0)=0 \\
H(a)=0 \\
Q(0)=0 \\
Q(b)=0 \\
Asi buscamos en nuestra tabla y vemos que nuestos eigenvalores estan definidas como:$(\frac{n\pi}{a})^2$ las eingenfuiones como $sin({n\pi}{a})$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Dimensiones
a = 4  # ft
b = 2  # ft

# Parámetros físicos
T = 12.5  # lb/ft
rho = 2.5  # slugs/ft^2

# Velocidad de onda
c = np.sqrt(T / rho)
print(f"Velocidad de propagación c = {c:.4f} ft/s")


def f(x, y):
    return 0.1 * (4*x - x**2) * (2*y - y**2)


def B_mn(m, n, a=4, b=2):
    x_vals = np.linspace(0, a, 500)
    y_vals = np.linspace(0, b, 500)
    X, Y = np.meshgrid(x_vals, y_vals)
    integrand = f(X, Y) * np.sin(m*np.pi*X/a) * np.sin(n*np.pi*Y/b)
    dx = a / (len(x_vals) - 1)
    dy = b / (len(y_vals) - 1)
    integral = np.sum(integrand) * dx * dy
    return (4/(a*b)) * integral

def u(x, y, t, N=5):
    total = 0
    for m in range(1, N+1):
        for n in range(1, N+1):
            B = B_mn(m, n)
            lambda_mn = c * np.pi * np.sqrt((m/a)**2 + (n/b)**2)
            total += B * np.cos(lambda_mn * t) * np.sin(m*np.pi*x/a) * np.sin(n*np.pi*y/b)
    return total

def plot_membrane_fast(t, N=5):
    x_vals = np.linspace(0, a, 80)
    y_vals = np.linspace(0, b, 80)
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = np.zeros_like(X)

    for m in range(1, N + 1):
        for n in range(1, N + 1):
            B = B_mn(m, n)
            lambda_mn = c * np.pi * np.sqrt((m / a)**2 + (n / b)**2)
            Z += B * np.cos(lambda_mn * t) * np.sin(m * np.pi * X / a) * np.sin(n * np.pi * Y / b)

    fig = plt.figure(figsize=(10, 6))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(X, Y, Z, cmap='plasma')
    ax.set_title(f"Desplazamiento de la membrana en t = {t:.2f} s")
    ax.set_xlabel("x (ft)")
    ax.set_ylabel("y (ft)")
    ax.set_zlabel("u(x, y, t)")
    plt.show()

plot_membrane_fast(0)    # Tiempo inicial
plot_membrane_fast(0.5)  # Medio segundo
plot_membrane_fast(1.0)  # Un segundo
```
