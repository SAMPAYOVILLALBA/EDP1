```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# El dominio espacial va de 0 a 1, dividido en 500 puntos.
#t_vals es una lista de tiempos en los que se evaluará la onda.
c = 1      # velocidad de propagación
L = 1      # longitud del dominio
x = np.linspace(0, L, 500)
t_vals = [0, 0.1, 0.2, 0.3, 0.4]

# función tipo Gauss
def gauss(x, t, direccion='+'):
    if direccion == '+':
        x_prime = x - c * t
    else:                             #Si la dirección es '+', se mueve hacia la derecha.  Si es '-', se mueve hacia la izquierda.
        x_prime = x + c * t
    return np.exp(-((x_prime - 0.5)**2) / 0.01)
#graficas
#f: la onda viajando a la derecha.
#g: la onda viajando a la izquierda.
#u: la suma de ambas (interferencia/superposición).
plt.figure(figsize=(12, 6))

for i, t in enumerate(t_vals):
    plt.subplot(2, 3, i+1)
    f = gauss(x, t, '+')
    g = gauss(x, t, '-')
    u = f + g

    plt.plot(x, f, 'y--', label='Derecha')
    plt.plot(x, g, 'd--', label='Izquierda')
    plt.plot(x, u, 'm', label='Superposición')
    plt.title(f"t = {t:.2f}")
    plt.ylim(0, 2.1)
    plt.grid(True)
    if i == 0:
        plt.legend()

plt.tight_layout()
plt.show()
```
