```python
#importando librerias
import numpy as np
import matplotlib.pyplot as plt
from sympy import symbols, diff, simplify, factorial, lambdify, Eq
from sympy.abc import x
from sympy.functions.special.polynomials import legendre
from scipy.special import legendre as scipy_legendre
# Verificar grficamente la identidad para n=3
n = 3 #se inicia n
Pn_expr = legendre(n, x) #la expresion de p_3
Pn1_expr = legendre(n + 1, x)#la expresion de p_4
lhs = (1 - x**2) * diff(Pn_expr, x) #la identidad del lado izquierdo
rhs = (n + 1) * x * Pn_expr - (n + 1) * Pn1_expr #la identidad del lado derechho
lhs_func = lambdify(x, lhs)
rhs_func = lambdify(x, rhs)

xg = np.linspace(-1, 1, 400)
lhs_vals = lhs_func(xg) #evaluan ambos lados de la identidad
rhs_vals = rhs_func(xg)

plt.figure(figsize=(10, 6)) #todo lo de la grafica
plt.plot(xg, lhs_vals, label='$(1 - x^2)P_3\'(x)$', linestyle='--',color='red' )
plt.plot(xg, rhs_vals, label='$(4xP_3(x) - 4P_4(x))$', linestyle=':',color='blue', linewidth=3.5) #
plt.title('Verificacin grfica de la identidad (b) para $n = 3$')
plt.xlabel('x')
plt.ylabel('Valor')
plt.axhline(0, color='gray', linewidth=0.5)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Parte (c) Verificacin algebraica de la identidad
left_c = simplify((1 - x**2) * diff(Pn_expr, x)) #se calcula el lado izquierdo de la identidad
right_c = simplify((n + 1) * x * Pn_expr - (n + 1) * Pn1_expr)
print("\n(c) Identidad (1 - x^2)P_3'(x) = 4xP_3(x) - 4P_4(x):", Eq(left_c, right_c)) #se imprime el resultado y se compaea con verdadero y falso

```

```python
# Parte (d) Verificacin algebraica de la identidad de recurrencia
Pn_1_expr = legendre(n - 1, x)
left_d = simplify((n + 1)*Pn1_expr - (2*n + 1)*x*Pn_expr + n*Pn_1_expr)
print("\n(d) Identidad de recurrencia:", Eq(left_d, 0))

```

```python
# Parte (e) Frmula de Rodrigues
print("\n(e) Verificacin de la frmula de Rodrigues:")
for k in range(6):
    R_expr = (1 / (2*k * factorial(k))) * diff((x**2 - 1)*k, x, k)
    Pk = legendre(k, x)
    correcto = simplify(R_expr - Pk) == 0
    print(f"P_{k}(x) =", R_expr, "==", Pk, "->", correcto)
```
