Considere la función pulso unitario: f(x)=1 si -1 <= x <= 1, 0 en otro caso.
a) Grafique esta función.
b) Calcule su representación como integral de Fourier, realizando específicamente las integrales en forma simbólica usando sympy. Grafique el resultado sobre la función original y compare.
c) Considere aproximaciones a la integral impropia, dando valores finitos al extremo superior de la integral: a= 8, 16 y 32 e integre simbólicamente en sympy. Gráfico sobre la función original en un panel distinto al del inciso b).
Ver páginas 61 y 62 del Kreyszig.
Tenemos la funcion \\
\
f(x) =
\begin{cases}
1, & \text{si } -1 \leq x \leq 1 \\
0, & \text{en otro caso}
\end{cases}
\
Ademas las siguientes formulas dadas en clase que son:
$\[
A(\omega) = \frac{1}{\pi} \int_{-\infty}^{\infty} f(u) \cos(\omega u)\, du
\quad , \quad \\
B(\omega) = \frac{1}{\pi} \int_{-\infty}^{\infty} f(u) \sin(\omega u)\, du
\]
\\
\[
f(x) = \int_0^\infty \left[ A(\omega) \cos(\omega x) + B(\omega) \sin(\omega x) \right] d\omega
\]$

```python

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from scipy.integrate import quad

#funcion de -1 a 1 dada el problema
x_vals = np.linspace(-3, 3, 400)
f_vals = np.where(np.abs(x_vals) <= 1, 1, 0)

plt.figure(figsize=(8, 3))
plt.plot(x_vals, f_vals, label='f(x) = 1 si |x| ≤ 1, 0 otro caso', color='blue')
plt.title("Función f(x) a trozos")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()

# Tenemos las serie de fourier dfadas las formulas en clase
x, t, w = sp.symbols('x t omega', real=True)
f_t = sp.Piecewise((1, (t >= -1) & (t <= 1)), (0, True))

# Cálculo s
a_w = sp.integrate(f_t * sp.cos(w * t), (t, -sp.oo, sp.oo))
b_w = sp.integrate(f_t * sp.sin(w * t), (t, -sp.oo, sp.oo))


# Resultado: a(ω) = 2 sin(ω)/pi ω, b(ω) = 0; comprobadfo en el libro

#  ω_max
def f_approx(x_array, w_max):
    """Devuelve f(x) aproximada mediante integral de Fourier con límite superior w_max"""
    result = []
    for x in x_array:
        integrand = lambda w: (2/np.pi) * (np.sin(w)/w) * np.cos(w * x)
        integral_val, _ = quad(integrand, 0, w_max)
        result.append(integral_val)
    return np.array(result)

# Graficas con las aproximaciones
plt.figure(figsize=(10, 5))
plt.plot(x_vals, f_vals, 'k--', label='f(x) original')

for w_max in [8, 16, 32]:
    approx_vals = f_approx(x_vals, w_max)
    plt.plot(x_vals, approx_vals, label=f'Aproximación con ω_max={w_max}')

plt.title("f(x) con los armonicos entregados")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
