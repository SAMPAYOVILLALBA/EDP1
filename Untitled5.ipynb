```python
from sympy import symbols, Function, Eq, dsolve, Derivative, sin, cos, pi, simplify, solve, exp
import matplotlib.pyplot as plt
import numpy as np

# Definimos símbolos
x, C1, C2, mu = symbols('x C1 C2 mu')
X = Function('X')
#Vamos a contemplar los casos para lambda
# Caso 1: lambda < 0
# Supongamos lambda = -mu**2, entonces la EDO: X'' - mu^2 X = 0, esto es resolver la ecuacion caracteristica
edo1 = Eq(Derivative(X(x), x, x) - mu**2 * X(x), 0)
sol1 = dsolve(edo1, X(x)).rhs

# Tenemos las condiciones de frontera: X(0)=0, X(1)=0 y sustituimos
eq1 = sol1.subs(x, 0)
eq2 = sol1.subs(x, 1)
conds1 = solve([eq1, eq2], (C1, C2))

#Caso 2: lambda = 0

edo2 = Eq(Derivative(X(x), x, x), 0)
sol2 = dsolve(edo2, X(x)).rhs

# Definimos otra ves las condiciones de frontera: X(0)=0, X(1)=0 para resolver estre caso
eq1_2 = sol2.subs(x, 0)
eq2_2 = sol2.subs(x, 1)
conds2 = solve([eq1_2, eq2_2], (C1, C2))

# Caso 3: lambda > 0
# Supongamos lambda = mu**2, entonces la EDO: X'' + mu^2 X = 0 para resolver mas facil la ecuacion carac.
edo3 = Eq(Derivative(X(x), x, x) + mu**2 * X(x), 0)
sol3 = dsolve(edo3, X(x)).rhs

# Aplicamos condiciones X(0)=0 => C1=0, X(1)=0 => sin(mu)=0 -> mu = n*pi
# Solución no trivial: C1=0, C2 arbitrario
# Eigenvalores: mu = n*pi => lambda_n = (n*pi)^2

# Graficamos algunas soluciones para lambda > 0
x_vals = np.linspace(0, 1, 400)
for n in range(1, 4):
    mu_val = n * np.pi
    y_vals = np.sin(mu_val * x_vals)
    plt.plot(x_vals, y_vals, label=f'n={n}, λ={mu_val**2:.2f}')
plt.title('Soluciones para λ > 0 (Dirichlet)')
plt.xlabel('x')
plt.ylabel('X(x)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
#En este grafico solo consideramos la solucion
```

```python
from sympy import symbols, Function, Eq, dsolve, Derivative, sin, cos, pi, simplify, solve, exp
import matplotlib.pyplot as plt
import numpy as np

x, C1, C2, mu = symbols('x C1 C2 mu')
X = Function('X')

# Dominio para graficar
x_vals = np.linspace(0, 1, 400)

# CASO 1: lambda < 0
# lambda = -mu^2 para que sea mas facil la ec, carc
mu_val = 3  # valor arbitrario positivo
sol1_expr = lambda x: np.sinh(mu_val * x) - np.sinh(mu_val * (1 - x))  # cumple las condiciones de frontera X(0)=X(1)=0
sol1_vals = sol1_expr(x_vals)

#  CASO 2: lambda = 0
# Solución: X(x) = C1 x + C2
sol2_expr = lambda x: 0 * x_vals
sol2_vals = sol2_expr(x_vals)

# CASO 3: lambda > 0
sol3_vals = []
labels = []
for n in range(1, 4):
    mu_n = n * np.pi
    sol = np.sin(mu_n * x_vals)
    sol3_vals.append(sol)
    labels.append(f'λ > 0: n={n}, λ={mu_n**2:.2f}')

# Graficar todos los casos juntos
plt.figure(figsize=(10, 6))
plt.plot(x_vals, sol1_vals, label='λ < 0 (μ=3)', linestyle='--')
plt.plot(x_vals, sol2_vals, label='λ = 0', linestyle=':')
for i, sol in enumerate(sol3_vals):
    plt.plot(x_vals, sol, label=labels[i])
plt.title('Soluciones de Sturm-Liouville para λ < 0, λ = 0, λ > 0 (Dirichlet)')
plt.xlabel('x')
plt.ylabel('X(x)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
#En esta garfica ya consideramos los 3 casos para lambda
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Definir variables
x = sp.symbols('x')
C1, C2, lam = sp.symbols('C1 C2 lambda')

## Caso 1: λ = 0
print("\nCaso 1: λ = 0")
X0 = C1 + C2*x
eq1_0 = sp.diff(X0, x).subs(x, 0)  # X'(0) = 0
eq2_0 = sp.diff(X0, x).subs(x, 1)  # X'(1) = 0

sol0 = sp.solve([eq1_0, eq2_0], (C1, C2), dict=True)
print("Solución general para λ=0:", X0)
print("Solución con condiciones de frontera:", sol0)
print("Solución no trivial: X(x) = constante (C1 arbitrario, C2=0)\n")

## Caso 2: λ < 0 (sea λ = -k^2)
print("\nCaso 2: λ < 0")
k = sp.symbols('k', positive=True)
X_neg = C1*sp.cosh(k*x) + C2*sp.sinh(k*x)  # Forma alternativa más estable

eq1_neg = sp.diff(X_neg, x).subs(x, 0)  # X'(0) = 0
eq2_neg = sp.diff(X_neg, x).subs(x, 1)  # X'(1) = 0

sol_neg = sp.solve([eq1_neg, eq2_neg], (C1, C2), dict=True)
print("Solución general para λ<0:", X_neg)
print("Solución con condiciones de frontera:", sol_neg)
print("Solución no trivial: Solo la solución trivial X(x)=0\n")

## Caso 3: λ > 0 (sea λ = k^2)
print("\nCaso 3: λ > 0 (Versión corregida)")
k = sp.symbols('k', positive=True)
X_pos = C1*sp.cos(k*x) + C2*sp.sin(k*x)

# Aplicamos condiciones de frontera
print("\nAplicando condiciones de frontera:")
print("1. X'(0) =", sp.diff(X_pos, x).subs(x, 0), "= 0 ⇒", sp.simplify(sp.diff(X_pos, x).subs(x, 0)), "= 0")
# Esto nos da directamente C2 = 0 (pues k ≠ 0)

# Sustituimos C2 = 0 en la solución
X_pos_reduced = X_pos.subs(C2, 0)
print("Solución reducida con C2=0:", X_pos_reduced)

print("\n2. X'(1) =", sp.diff(X_pos_reduced, x).subs(x, 1), "= 0")

print("Para solución no trivial (C1 ≠ 0), requerimos sin(k) = 0 ⇒ k = nπ, n ∈ ℤ+")

# Autovalores y autofunciones
n = sp.symbols('n', integer=True, positive=True)
lambda_n = (n*sp.pi)**2
X_n = sp.cos(n*sp.pi*x)

print("\nResultados finales:")
print("Autovalores λ_n:", lambda_n)
print("Autofunciones X_n(x):", X_n)

# Graficar las autofunciones no triviales
print("\nGraficando autofunciones...")
x_vals = np.linspace(0, 1, 100)
plt.figure(figsize=(10, 6))

for n_val in range(1, 5):
    X_n_func = sp.lambdify(x, X_n.subs(n, n_val), 'numpy')
    y_vals = X_n_func(x_vals)
    plt.plot(x_vals, y_vals, label=f'n={n_val} (λ={(n_val*np.pi)**2:.2f})')

plt.title("Autofunciones para X'' + λX = 0 con X'(0)=X'(1)=0")
plt.xlabel("x")
plt.ylabel("X(x)")
plt.legend()
plt.grid(True)
plt.show()
```
