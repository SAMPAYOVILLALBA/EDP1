```python
import numpy as np
import matplotlib.pyplot as plt

def u(x, N=20):
    f = np.zeros(len(x))
    g = np.zeros(len(x))
    for n in range(1, N):
        g += (2 * (1 + (-1) ** (n + 1)) / (n * np.pi)) * np.sin(n * np.pi * x / 2)
    f = 1/2 + g
    return f


x1 = np.linspace(-2, 0, 200)
x2 = np.linspace(0, 2, 200)
x3 = np.linspace(-2, 2, 200)


y1 = np.zeros(len(x1))
y2 = 2 - x2
y3 = u(x3, N=200)


plt.title("Gráfica de f(x)")

plt.plot(x1, y1, label='y=0')
plt.plot(x2, y2, label='y=2-x')
plt.plot(x3, y3, label='u=(N=200)')

for i in range(5, 20, 5):
    plt.plot(x3, u(x3, i), label=f'$f_N(x)$ con N={i}')

plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title("Variación de Armónicos")

plt.legend()
plt.grid(True)
plt.show()
```
