```python
#EJERCICIO 1
#Con el ejemplo de la aproximación de PI ¿Cuántos dardos hay que lanzar para asegurar que (P|X_n-PI/4|>0.01)<0.1?
from random import random
from math import pi
import matplotlib.pyplot as plt
import numpy as np
Nc=0 # Número de dardos que entra en la aproximación de PI
N=10000 # Número de veces que se lanza el dado
lx=[] #eje X
ly=[] #eje de la circunferencia PI
lc=[] #valor de pi
Xn=[]  #1 si el dardo cae dentro del circilo, 0 en otro caso
for i in range(N):
  x=random()
  y=random()
  if x**2+y**2<1:
    Nc+=1
    XN=1   #El dardo entra en el círculo
    Xn.append(XN)
  else:
    XN=0 #El dardo no entra en el círculo
    Xn.append(XN)
  lx.append(i+1)
  ly.append(4*(Nc/(i+1)))   # 4 veces los dardos que entran en el circulo /la cantidad de dardos que fueron lanzados
  lc.append(pi)
Ex=1/2
Varx=Ex*(1-Ex) #/N
N1=((0.01**2)*0.1)/Varx
print("E(x)=",Ex) #Probabilidad de que el dardo caiga en el círculo
print("Var(X)",Varx)
print("La aproximación de pi es ",4*(Nc/N))
print(N1)
#ERROR MAYOR DE 0.001 SEA MENOR QUE 0.1
#P(|Xn-PI|>0.01)<0.1   E=0.01
#DESIGUALDAD DE CHEBYSHEV P(|Z-E(Z)|>E)<=V(Z)/E^2
#p(1-p)/(E^2)(Sea menor que)<N
plt.plot(lx,ly)
plt.plot(lx,lc)
plt.show()
```

```python
#EJERCICIO 2
# Empleando un método similar al empleado en esta sección para aproximar π, estime mediante simulación el valor de √2.
from random import random
import matplotlib.pyplot as plt
import numpy as np
from math import sqrt
def raiz():
  N=1000 #Numero de veces que se lanza el dardo (raiz de dos esta entre 1 y 2)
  n=0
  lx=[]
  ly=[]
  lc=[]
  for i in range(N):
    x=1+random()
    if x**2<2:
      n+=1
    lx.append(i+1)
    ly.append(1+n/(i+1)) #es lo que se le agrega la random/(1+i)
    lc.append(sqrt(2))
  plt.plot(lx,ly)
  plt.plot(lx,lc)
  plt.show()
  return 1+n/N
print("la aproximación de raiz de dos es",raiz())
```

```python
#EJERCICIO 3
#Estime mediante simulación el volumen de una esfera de radio 1
from random import random
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt, log, exp, pi
def esfera(): #V=4/3PIR^3
  N=100000
  n1=0
  Nc=[]  #cuando cae dentro de la esfera
  lx=[]
  ly=[]
  lc=[]
  for i in range(N):
    x=-1+2*random()
    y=-1+2*random()
    z=-1+2*random()
    if x**2+y**2+z**2<1:
      n=1
      n1+=1
      Nc.append(n)
    else:
      n=0
      Nc.append(n)
    lx.append(i+1)
    ly.append(8*n1/(i+1))
    lc.append((4/3)*pi)
  plt.plot(lx,ly)
  plt.plot(lx,lc)
  plt.show()
  print(Nc)
  return 8*sum(Nc)/N
print("El volumen aproximado de la esfera es", esfera())
```

```python
#EJERCICIO 4
# Dos jugadores A y B lanzan volados hasta que alguno de los dos se queda sin dinero. Si A gana, B paga un peso a A; si B gana, A paga un peso a B. A tiene
# inicialmente 3 pesos y B tiene 12 pesos. Estime mediante simulación la probabilidad de que A gane al final del juego y el valor esperado
# del número de volados que se realizarán en el juego.
from random import random
import numpy as np
import matplotlib.pyplot as plt
def volados():
  A=3
  B=12
  lanzamientos=0
  while A>0 and B>0:
    x=random()
    lanzamientos+=1
    if x<0.5:
      A+=1
      B-=1
    else:
      A-=1
      B+=1
  return A,B,lanzamientos
N=100000
ganaA=0
lanzamientos_xronda=[]
for i in range(N):
  A,B,lanzamientos=volados()
  if B==0:
    ganaA+=1
  lanzamientos_xronda.append(lanzamientos)
print("P(A gane)=",ganaA/N)
print("E(lanzamientos)=",np.mean(lanzamientos_xronda))
```

```python
#EJERCICIO 7
#Estimar el valor de las integrales de 0 a 1 de 1/(1+x^2)dx y de 0 a 1 de raiz(1-x^2)dx usando simulación
from random import random
from math import sqrt
def integral1():
  x=random()
  fx1=1/(1+x**2)
  return fx1
def integral2():
  x=random()
  fx2=sqrt(1-x**2)
  return fx2
N=100000
sumaInt1=0
sumaInt2=0
for i in range(N):
  sumaInt1+=integral1()
  sumaInt2+=integral2()
print("La aproximación de la integral 1/(1+x^2) de 0 a 1 es", sumaInt1/N)
print("La aproximación de la integral (1-x^2)^(1/2) de 0 a 1 es", sumaInt2/N)
```

```python
#EJERCICIO 9
#Las dos integrales del problema 7 tienen el valor pi/4 ¿Cuál de las dos integrales del problema anterior produce estimaciones con menor varianza?
from random import random
from math import sqrt,pi
import numpy as np
def integral1():
  x=random()
  fx1=1/(1+x**2)
  return fx1
def integral2():
  x=random()
  fx2=sqrt(1-x**2)
  return fx2
N=100000
sumaInt1=0
var1=[]
sumaInt2=0
var2=[]
for i in range(N):
  sumaInt1+=integral1()
  var1.append((integral1()-(pi/4))**2)
  sumaInt2+=integral2()
  var2.append((integral2()-(pi/4))**2)
print("La aproximación de la integral 1/(1+x^2) de 0 a 1 es", sumaInt1/N)
print("La aproximación de la integral (1-x^2)^(1/2) de 0 a 1 es", sumaInt2/N)
print("La varianza de la integral 1 es",sum(var1)/N)
print("La varianza de la integral 2 es",sum(var2)/N)
```

```python
#EJERCICIO 10
#Estime el valor de 0 a ∞ de e^(-x^2)dx .Encuentre un intervalo de confianza del 90% para la estimación del valor de la integral
from random import random
from math import sqrt,pi, exp
import numpy as np
def integral():
  u=random()
  x=u/(1-u)
  fx=exp(-x**2)/(1-u)**2
  return fx
N=10000
I=[]
for i in range(N):
  I.append(integral())
desvest=sqrt(np.var(I))/sqrt(N)
Z=1.645 #Valor de Z al 90%
print("el valor de la integral es", np.mean(I))
print("Desviación Estándar",desvest)
print("IC_90%=[",np.mean(I)-Z*desvest,",",np.mean(I)+Z*desvest,"]")
```

```python
#EJERCICIO 11
# Por medio de simulación estimar el valor de la doble integral de 0 a ∞ de e^{-x^2+y^2}dxdy. Encuentre un intervalo de confianza del
# 90% para la estimación del valor de la integral
from random import random
from math import sqrt, log, exp
import numpy as np
def integral():
  u1=random()
  u2=random()
  x=u1/(1-u1)
  y=u2/(1-u2)
  u=1/((1-u1)**2*(1-u2)**2) #Jacobiano
  fx=exp(-(x**2+y**2))*u
  return fx
N=10000
l=[]
for i in range(N):
  l.append(integral())
esperado=np.mean(l)
varianza=np.var(l)
desvest=sqrt(varianza/N)
Z=1.645
print("El estimado de la integral es",esperado)
print("IC_95%=[",esperado-Z*desvest,",",esperado+Z*desvest,"]")
```

```python
#EJERCICIO 13
#Generar una muestra de 1 000 000 números aleatorios para una variable que siga una distribución binomial con parámetros p=0.75 y n=7
#Calcular la media y la varianza de la muestra y compararla con los valores np y np(1-p)
from random import random
import numpy as np
n=7
p=0.75
def binomial(n,p):
  u=random()
  pi=(1-p)**n
  s=pi
  k=0
  while s<u:
    pi=(pi*(n-k)/(k+1))*(p/(1-p))
    s+=pi
    k+=1
  return k
l=[]
N=1000000
for i in range(N):
  x=binomial(n,p)
  l.append(x)
media=np.mean(l)
varianza=np.var(l)
print("Media por simulación",media)
print("Media teoríca",n*p)
print("varianza por simulación",varianza)
print("varianza teorica",n*p*(1-p))
```

```python
#EJERCICIO 14
#Genera una muestra aleatoria para una variable que sigue un adistribución de Poisson con parámetro λ=5 y calcular la media y la varianza de
#la muestra y compararlas con λ
from random import random
import numpy as np
def poisson():
  u=random()
  l=5
  p=exp(-l)
  s=p
  k=0
  while s<u:
    p=p*(l/(k+1))
    s+=p
    k+=1
  return k
N=10000
muestra=[]
for i in range(N):
  x=poisson()
  muestra.append(x)
media=np.mean(muestra)
varianza=np.var(muestra)
print("media por simulación",media)
print("media teorica",5)
print("varianza por simulación",varianza)
print("varianza teorica", 5)
```

```python
#EJERCICIO 15
# Generar una muestra aleatoria de tamaño 1 000 000 para una variable que siga una distribución geométrica con parámetro p=0.25,
# calcular la media y la varianza y comparar con los valores media= 1/p y varianza= (1-p)/p^2
from random import random
import numpy as np
from math import log
def geometrica():
  u=random()
  p=0.25
  x=int(log(1-u)/log(1-p))+1
  return x
N=1000000
muestra=[]
for i in range(N):
  x=geometrica()
  muestra.append(x)
media=np.mean(muestra)
varianza=np.var(muestra)
p=0.25
print("la media por simulación es",media)
print("la media teorica es", 1/p)
print("la varianza por simulación es", varianza)
print("la varianza teorica es", (1-p)/p**2)
```

```python
#EJERCICIO 16
# La variable aleatoria X tiene a la siguiete función de densidad de probabilidad P(X=i)=4/5^i para i=1,2,...
# Estime por simulación el valor esperado de X^{1/2} pasar de densidad a Acumulativa (f a F)
from random import random
from math import sqrt
import numpy as np
def fun_densidad():
  u=random()
  i=1
  p=4/5
  s=p
  while s<u:
    p=4/(5**(i+1))
    i+=1
    s+=p
  return i
N=1000000
l=[]
for j in range(N):
  x=fun_densidad()
  l.append(sqrt(x))
media=np.mean(l)
print("el valor esperado de X^{1/2}",media)
```
