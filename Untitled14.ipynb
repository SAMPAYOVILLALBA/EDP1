```python
!pip install yfinance --quiet

import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime, timedelta

%matplotlib inline

# ================================================================
#  PARMETROS GENERALES
# ================================================================
ticker = "FIBRAMQ12.MX"   # FIBRA Macquarie Mxico
hoy = datetime.today()
inicio = hoy - timedelta(days=5*365)

print(f"Descargando datos de {ticker} desde {inicio.date()} hasta {hoy.date()}...\n")

# ================================================================
#  DESCARGA DE PRECIOS
# ================================================================
precios = yf.download(ticker, start=inicio, end=hoy)

print("Primeras filas de la serie de precios:")
display(precios.head())

# ================================================================
#  OBTENER SERIE 'CLOSE' ROBUSTA (POR SI VIENE EN MULTIINDEX)
# ================================================================
if isinstance(precios.columns, pd.MultiIndex):
    # Nivel 0 = tipo de precio, nivel 1 = ticker
    close_series = precios["Close"].iloc[:, 0]
else:
    close_series = precios["Close"]

# ================================================================
#  GRFICA DEL PRECIO (Close)
# ================================================================
plt.figure(figsize=(10, 4))
plt.plot(precios.index, close_series, color="red")
plt.title("Fibra Macquarie Mxico (FIBRAMQ12.MX)  Precio Close ltimos 5 aos")
plt.xlabel("Fecha")
plt.ylabel("Precio (MXN)")
plt.grid(True)
plt.show()

# ================================================================
#  DIVIDENDOS (CONVERSIN Y TZ CORREGIDAS)
# ================================================================
fibra = yf.Ticker(ticker)
divs = fibra.dividends   # Serie: ndice = fecha, valores = dividendo

# convertir ndice a datetime
divs.index = pd.to_datetime(divs.index)

# quitar zona horaria
divs.index = divs.index.tz_localize(None)

# filtrar dividendos a ltimos 5 aos
divs = divs[divs.index >= inicio]

print("\nDividendos encontrados (ltimos 5 aos):")
display(divs)

# ================================================================
#  RESUMEN ANUAL Y GRFICA DE DIVIDENDOS
# ================================================================
if divs.empty:
    print("\n Esta FIBRA no tiene dividendos registrados en Yahoo Finance en los ltimos 5 aos.")
    divs_anuales = pd.DataFrame()
else:
    # agrupar por ao (YE = Year End)
    divs_anuales = divs.resample("YE").sum()
    divs_anuales.index = divs_anuales.index.year

    print("\nDividendos totales por ao:")
    display(divs_anuales)

    # grfica de dividendos anuales
    plt.figure(figsize=(8, 4))
    plt.bar(divs_anuales.index.astype(str), divs_anuales.values)
    plt.title("Fibra Macquarie Mxico  Dividendos anuales (ltimos 5 aos)")
    plt.xlabel("Ao")
    plt.ylabel("MXN por ttulo (suma anual)")
    plt.grid(axis="y")
    plt.show()

# ================================================================
#  CLCULO DE DIVIDENDO PROMEDIO Y TABLA DE FLUJOS
# ================================================================
if not divs.empty:
    dividendo_promedio = divs_anuales.mean().item()
    print(f"\nDividendo anual promedio (ltimos 5 aos): {dividendo_promedio:.2f} MXN por ttulo")

    anios_proyeccion = 5
    flujos = [dividendo_promedio] * anios_proyeccion

    tabla_flujos = pd.DataFrame({
        "Ao": list(range(1, anios_proyeccion + 1)),
        "Flujo_dividendo_estimado": flujos
    })

    print("\nTabla de flujos estimados (como si fuera un bono):")
    display(tabla_flujos)
else:
    print("\nNo se puede calcular dividendo promedio ni flujos porque no hay datos de dividendos.")

# ================================================================
#  VALUACIN TIPO BONO
# ================================================================
tasa_descuento = 0.10  # 10% de ejemplo, puedes cambiarlo por CETES

# ltimo precio de cierre como float
precio_actual = float(close_series.iloc[-1])
print(f"\nPrecio actual de mercado: {precio_actual:.2f} MXN")

if not divs.empty:
    anios_proyeccion = 5
    flujos = [dividendo_promedio] * anios_proyeccion
    precio_final_estimado = precio_actual  # suponemos que en 5 aos vale lo mismo

    # valor presente de los dividendos
    vp_dividendos = sum(cf / (1 + tasa_descuento)**t for t, cf in enumerate(flujos, start=1))
    # valor presente del precio final
    vp_precio_final = precio_final_estimado / (1 + tasa_descuento)**anios_proyeccion

    valor_teorico = vp_dividendos + vp_precio_final

    print(f"\nValor presente de los dividendos: {vp_dividendos:.2f} MXN")
    print(f"Valor presente del precio futuro: {vp_precio_final:.2f} MXN")
    print(f"\n Valor terico (modelo tipo bono): {valor_teorico:.2f} MXN")
    print(f" Precio de mercado: {precio_actual:.2f} MXN\n")

    if valor_teorico > precio_actual:
        print(" Interpretacin: La FIBRA podra estar SUBVALUADA (conviene comprar).")
    else:
        print(" Interpretacin: La FIBRA podra estar SOBREVALUADA o en precio.")
else:
    print("\n No se puede hacer la valuacin tipo bono porque no hay dividendos registrados.")
# Descargar datos



# =====================================================
#  CREAR EXCEL CON PRECIOS + DIVIDENDOS POR FECHA
# =====================================================

# Asegurar que los ndices estn en datetime sin tz
precios.index = pd.to_datetime(precios.index)
divs.index = pd.to_datetime(divs.index)

# Crear DataFrame con fechas y precios
df_export = precios.copy()

# Agregar la columna de dividendos usando reindex (Rellena con 0 si no hubo)
df_export['Dividendos'] = divs.reindex(df_export.index).fillna(0)

# Exportar a Excel
df_export.to_excel("FIBRAMQ12_precios_dividendos.xlsx", index=True)

print("\nArchivo generado: FIBRAMQ12_precios_dividendos.xlsx")

# Descargar archivo en Colab
from google.colab import files
files.download("FIBRAMQ12_precios_dividendos.xlsx")
```
